```python
from piml import Experiment
exp = Experiment()
```

```python
import pandas as pd
data = pd.read_csv("https://github.com/SelfExplainML/PiML-Toolbox/blob/main/datasets/SimuCredit_v2.csv?raw=true")
exp.data_loader(data=data, silent=True)
exp.data_summary(feature_exclude=["Gender", "Race"], silent=True)
exp.data_prepare(test_ratio=0.2, random_state=0, silent=True)
```

```python
# Choose XGB2 (default), run;
# then choose XGB2-Mono and GAMI-Net with Feature_increasing: "Balance",
exp.model_train()
```

```python
# Choose XGB2, Check Effect Plot
exp.model_interpret()
```

```python
# Choose XGB2-Mono, Check Effect Plot
exp.model_interpret()
```
